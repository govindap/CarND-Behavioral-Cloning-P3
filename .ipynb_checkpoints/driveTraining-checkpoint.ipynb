```python
from __future__ import absolute_import
from __future__ import print_function

import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
%matplotlib inline
import cv2
from keras.applications.resnet50 import ResNet50, preprocess_input
from keras.applications.inception_v3 import InceptionV3
from keras.applications.vgg16 import VGG16
from keras.layers import Input, AveragePooling2D, Dense, Flatten
from sklearn.model_selection import train_test_split
from keras.datasets import cifar10
import pickle
import tensorflow as tf
import keras.backend as K
from keras.models import Sequential, Model
from keras.layers import Lambda
from keras.layers import Cropping2D
from keras.models import Sequential, Model
from keras.layers.core import Dense, Dropout, Activation, Flatten, Reshape
from keras.layers import BatchNormalization,Input
from keras.layers.recurrent import SimpleRNN, LSTM
from keras.layers.convolutional import Convolution2D
from keras.optimizers import SGD, Adam, RMSprop
import sklearn.metrics as metrics
from PIL import Image
import csv
from keras.callbacks import ModelCheckpoint,EarlyStopping,History
```

```python
imgdir = r'F:/downloads/simulator-windows-64/data/data/'
imgdir1 = r'F:/downloads/simulator-windows-64/data1/driving_log.csv'
imgdir2 = r'F:/downloads/simulator-windows-64/data2/driving_log.csv'
imgdir3 = r'F:/downloads/simulator-windows-64/data3/driving_log.csv'
imgdir4 = r'F:/downloads/simulator-windows-64/data4/driving_log.csv'
csvlist = [imgdir+r'driving_log.csv',imgdir1,imgdir2,imgdir3,imgdir4]
dirlist = [imgdir,'','','','']
```

```python
car_images = []
steering_angles = []
# To capture flipped images sample from dataset to verify it
fimages = []
# To capture center, left, right sample images from dataset to verify it
simages = []
for csv_file,path in zip(csvlist,dirlist):
    with open(csv_file, 'r') as f:
        reader = csv.reader(f)
        next(reader)
        for row in reader:
            steering_center = float(row[3].strip())

            # create adjusted steering measurements for the side camera images
            correction = 0.15 # this is a parameter to tune
            steering_left = steering_center + correction
            steering_right = steering_center - correction
            
            # read in images from center, left and right cameras
            img_center = np.asarray(Image.open(path + row[0].strip()))
            img_left = np.asarray(Image.open(path + row[1].strip()))
            img_right = np.asarray(Image.open(path + row[2].strip()))
            
            if(len(simages) < 7 and steering_center >0.3):
                simages.append(img_center)
                simages.append(img_left)
                simages.append(img_right)
            
            if(steering_center>0.3 or steering_center<-0.3):
                measurement_flipped = -steering_center
                image_flipped = np.fliplr(img_center)
                #image_flipped = cv2.flip(img_center, 1)
                car_images.append(image_flipped)
                steering_angles.append(measurement_flipped)
                if(len(fimages) < 5):
                    fimages.append(img_center)
                    fimages.append(image_flipped)
                
            # add images and angles to data set
            car_images.extend([img_center, img_left, img_right])
            steering_angles.extend([steering_center, steering_left, steering_right])

car_images = np.array(car_images)
steering_angles = np.array(steering_angles)
```

```python
print(car_images.shape)
print(steering_angles.shape)
```

```python
ch, row, col = 3, 90, 320  # Trimmed image format
nrows = row
ncols = col
#NVIDIA model
def create_model():
    model = Sequential()
    model.add(Cropping2D(cropping=((50,20), (0,0)), input_shape=(160,320,3)))    
    model.add(Lambda(lambda x: x/127.5 - 1.,input_shape=(row, col,ch),output_shape=(row, col,ch)))    
    model.add(BatchNormalization(epsilon=0.001,mode=2, axis=1,input_shape=(nrows,ncols,3)))
    model.add(Convolution2D(24,5,5,border_mode='valid', activation='relu', subsample=(2,2)))
    model.add(Convolution2D(36,5,5,border_mode='valid', activation='relu', subsample=(2,2)))
    model.add(Convolution2D(48,5,5,border_mode='valid', activation='relu', subsample=(2,2)))
    model.add(Convolution2D(64,3,3,border_mode='valid', activation='relu', subsample=(1,1)))
    model.add(Convolution2D(64,3,3,border_mode='valid', activation='relu', subsample=(1,1)))
    model.add(Flatten())
    model.add(Dense(1164, activation='relu'))
    model.add(Dense(100, activation='relu'))
    model.add(Dense(50, activation='relu'))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(1, activation='tanh'))
    return model


```

```python
nb_epoch = 10
batch_size = 256

# checkpoint
filepath="weights.best.hdf5"
checkpoint = ModelCheckpoint(filepath, monitor='val_acc', verbose=1, save_best_only=True, mode='max')
es = EarlyStopping(monitor='val_loss', min_delta=0, patience=2, verbose=0, mode='auto')
hs = History()
callbacks_list = [checkpoint,es,hs]

model = create_model()
model.compile(loss='mse', optimizer='adam',metrics=['mse','accuracy'])

with tf.device('/gpu:0'):
    history = model.fit(car_images, steering_angles, callbacks=callbacks_list,
    batch_size =batch_size, nb_epoch=nb_epoch, verbose=1,
    validation_split=0.1,shuffle=True)

model.save_weights('model_basic_weight.hdf5')
```

```python
import json
model_json = model.to_json()
with open("./drive2_model.json", "w") as json_file:
    json.dump(model_json, json_file)
model.save('drive2_model.h5') 
print("Saved model to disk")
```

```python
history_object = history

### print the keys contained in the history object
print(history_object.history.keys())

### plot the training and validation loss for each epoch
plt.plot(history_object.history['loss'])
plt.plot(history_object.history['val_loss'])
plt.title('model mean squared error loss')
plt.ylabel('mean squared error loss')
plt.xlabel('epoch')
plt.legend(['training set', 'validation set'], loc='upper right')
plt.show()
```
